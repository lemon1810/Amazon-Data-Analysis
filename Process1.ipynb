```python
import numpy as np
import pandas as pd
import re
```

#Process drop ID

```python
path='/content/gdrive/MyDrive/IE224/data_final/data.xlsx'
```

```python
df= pd.read_excel(path)
```

```python
df.columns
```

```python
df.info()
```

```python
df = df.drop(columns=['ID'])
```

# Process RAM

```python
df['Ram_capacity'] = df['Ram'].apply(lambda x: ''.join(filter(str.isdigit, str(x))))
df['Ram_model'] = df['Ram'].apply(lambda x: ''.join(filter(str.isalpha, str(x))))
```

```python
df[['Ram_capacity', 'Ram_model']].value_counts()
```

```python
# Assuming df is your DataFrame and 'CPU' is the column you want to process
df['Ram_model'] = df['Ram_model'].str.replace('GB', '')
df['Ram_model'] = df['Ram_model'].str.replace(' ', '')
df['Ram_capacity'] = df['Ram_capacity'].str.replace(' ', '')
df['Ram_capacity'] = df['Ram_capacity'].str.replace('nan', '')

# Display the DataFrame
df[['Ram_capacity', 'Ram_model']].value_counts()
```

# Process Screen

```python
import pandas as pd

# Assuming df is your DataFrame and 'screen' is the column you want to split
df[['Touchscreen', 'Screen_resolution']] = df['ScreenResolution'].str.extract(r'(Touchscreen)?\s?(\d+x\d+)')
# Convert columns to numeric type
df['Touchscreen'] = df['Touchscreen'].notnull().astype(int)
```

```python
df[['Touchscreen', 'Screen_resolution']].value_counts()
```

# Process CPU

```python
df[['Brand_RAM', 'Model', 'Clockspeed']] = df['Cpu'].str.extract(r'(\S+)\s(.*?)\s(\S+)$')
```

```python
df['Clockspeed'] = df['Clockspeed'].str.replace('GHz', '')
```

```python
df['Clockspeed'].value_counts()
```

```python
df['Model'].value_counts()
```

# Memory ->

```python
df['Memory'] = df['Memory'].str.replace('Memory Size ', '')
df['Memory'] = df['Memory'].str.replace('Memory size ', '')
```

```python
df[['Memory1', 'Memory2']] = df['Memory'].str.split(' \+ ', expand=True)
```

```python
df['Memory1'].value_counts()
```

```python
df['Memory2'].value_counts()
```

```python
df['Capacity1'] = df['Memory1'].str.extract(r'(\d+\s?(?:GB|TB)?)')
df['Storage_Type_1'] = df['Memory1'].str.extract(r'(SSD|HDD|Flash Storage|Hybrid)')

df['Capacity2'] = df['Memory2'].str.extract(r'(\d+\s?(?:GB|TB)?)')
df['Storage_Type_2'] = df['Memory2'].str.extract(r'(SSD|HDD|Flash Storage|Hybrid)')
```

```python
df['Capacity1']= df['Capacity1'].str.replace('TB', '000')
df['Capacity2']= df['Capacity2'].str.replace('TB', '000')
```

```python
df['Capacity1']= df['Capacity1'].str.replace(' ', '')
df['Storage_Type_1']= df['Storage_Type_1'].str.replace(' ', '')
df['Capacity2']= df['Capacity2'].str.replace(' ', '')
df['Storage_Type_2']= df['Storage_Type_2'].str.replace(' ', '')
```

```python
df['Capacity1']= df['Capacity1'].str.replace('GB', '')
df['Capacity2']= df['Capacity2'].str.replace('GB', '')
```

```python
df['Capacity1']= df['Capacity1'].str.replace('GB', '')
df['Capacity2']= df['Capacity2'].str.replace('GB', '')
```

```python
df['Capacity1'].value_counts()
```

```python
df['Capacity2'].value_counts()
```

```python
df['Storage_Type_1'].value_counts()
```

```python
df['Storage_Type_2'].value_counts()
```

#Process Weight
